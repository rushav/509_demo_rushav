# FizzBuzz
Write a program that prints the numbers from 1 to 100. But for multiples of three print “Fizz” instead of the number and for the multiples of five print “Buzz”. For numbers that are multiples of both three and five print “FizzBuzz”.

````python
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
````

# Tic-Tac-Toe Game Winner
Given a board of a tic-tac-toe game, find the winner. A sample board is given as follows:
```py
board = [['O', 'X', 'X'],
         ['X', 'X', 'O'],
         ['O', 'X', 'X']]
```


````python
board = [['O', 'X', 'X'],
         ['X', 'X', 'O'],
         ['O', 'X', 'X']]

winner = None

# Check all rows one by one
for row in board:
    if row[0] == row[1] == row[2]:
        winner = row[0]

# Check all columns one by one
for col in range(3):
    if board[0][col] == board[1][col] == board[2][col]:
        winner = board[0][col]

# Check diagonal (top-left to bottom-right)
if board[0][0] == board[1][1] == board[2][2]:
    winner = board[0][0]

# Check diagonal (top-right to bottom-left)
if board[0][2] == board[1][2] == board[2][0]:
    winner = board[0][2]

if winner:
    print(f"Winner is: {winner}")
else:
    print("No winner")
````

# Password Strenghth Meter
Write a Python program that reads passwords from users. For the given password, tell users whether the password is **weak**, **okay**, or **strong**. Here are the scoring rules for a password:

- Length
  - +2 if length ≥ 12
  - +1 if length is 8–11
  - +0 otherwise
- Character variety
  - +1 if it has both lowercase [a-z] and uppercase [A-Z]
  - +1 if it has at least one digit [0-9]
  - +1 if it has at least one special from this set only: !@#$%^&*
- Penalties
  - −2 if it contains any whitespace (space, tab, etc.)
  - −3 if it contains three or more identical characters in a row (e.g., aaa, 1111, ***)
- Final Result
  - weak if score < 3
  - okay if 3 ≤ score ≤ 4
  - strong if score ≥ 5

Test your program with the following passwords:
- P@ssw0rd
- hunter2
- CorrectHorseBatteryStaple
- abcde Fgh
- A!!!A!!!A!!!

**Hint**: You may find `str.islower()`, `str.isupper()`, `str.isdigit()`, and `str.isspace()` useful.

````python
password = input("Enter password: ")

# Initialize all variables
score = 0
has_lower = False
has_upper = False
has_digit = False
has_special = False
has_whitespace = False
has_three_identical = False

# Check for length
if len(password) >= 12:
    score += 2
elif len(password) >= 8:
    score += 1

# Single loop to check all character properties
special_chars = "!@#$%^&*"
for char in password:
    if char.islower():
        has_lower = True
    
    if char.isupper():
        has_upper = True
    
    if char.isdigit():
        has_digit = True
    
    if char in special_chars:
        has_special = True
    
    if char.isspace():
        has_whitespace = True

# Calculate score based on findings
if has_lower and has_upper:
    score += 1

if has_digit:
    score += 1

if has_special:
    score += 1

if has_whitespace:
    score -= 2

# Check for three identical characters in a row
for i in range(len(password) - 2):
    if password[i] == password[i+1] == password[i+2]:
        has_three_identical = True

if has_three_identical:
    score -= 3

# Determine strength
if score < 3:
    strength = "weak"
elif score <= 4:
    strength = "okay"
else:
    strength = "strong"

print(f"Password strength: {strength} (score: {score})")
````

# Vacuum Robot

You get an $n\times n$ map made of:
- \# = wall (can’t go through)
- . = empty floor (can move)
- S = where the robot starts (treated like empty floor)
You’re told which way the robot is facing (N, E, S, or W).

You get a string of commands:
- L turn left 90°
- R turn right 90°
- F try to move one step forward
If a forward step would hit a wall or leave the map, the robot remains where it is. Its health condition decreases by 1.
After all commands, print: row col orientation (rows/cols are 1-based, top-left is (1,1)), as well as robot's health condition.
**Note**: The minimum health condition is 0. Once the robot's health reaches 0, it cannot accept commands any more. 

Test your program with the following grids and commands:

- Grid 1
    ```py
    #####
    #...#
    #.#S#
    #...#
    #####
    ```
    Commands: FFLFFRFFF; assume the robot initially facing east (E) with initial health 3.

- Grid 2
    ```py
    #######
    #..#..#
    #.#.#S#
    #.#...#
    #..####
    #.....#
    #######
    ```
    Commands: FLLFFFRFFLFFRFF; assume the robot initially facing west (W) with initial health 10.

````python
# Initial grid and parameters (based on Grid 1)
grid = [
    "#####",
    "#...#",
    "#.#S#",
    "#...#",
    "#####"
]
commands = "FFLFFRFFF"
facing = "E"
health = 3

# Find starting position
start_row = 0
start_col = 0
for i in range(len(grid)):
    for j in range(len(grid[i])):
        if grid[i][j] == 'S':
            start_row = i + 1 
            start_col = j + 1

# Current position
row = start_row
col = start_col

# Direction mappings
directions = ['N', 'E', 'S', 'W']
current_dir = directions.index(facing)

# Process each command
for command in commands:
    if health == 0:
        break
    
    if command == 'L':
        # Turn left
        current_dir = (current_dir - 1) % 4
    
    elif command == 'R':
        # Turn right
        current_dir = (current_dir + 1) % 4
    
    elif command == 'F':
        # Try to move forward
        new_row = row
        new_col = col
        
        # Calculate new position based on the recent direction
        # North means move up a row (decrease row index)
        # East means move right a column (increase column index)
        # South means move down a row (increase row index)
        # West means move left a column (decrease column index)
        if directions[current_dir] == 'N':
            new_row = row - 1
        elif directions[current_dir] == 'E':
            new_col = col + 1
        elif directions[current_dir] == 'S':
            new_row = row + 1
        elif directions[current_dir] == 'W':
            new_col = col - 1
        
        # Check if new position is valid 
        # Collision detection
        if new_row < 1 or new_row > len(grid) or new_col < 1 or new_col > len(grid[0]):
            # Out of bounds
            health -= 1
            if health < 0:
                health = 0
        elif grid[new_row - 1][new_col - 1] == '#':
            # Hit a wall
            health -= 1
            if health < 0:
                health = 0
        else:
            # Valid move
            row = new_row
            col = new_col

print(f"{row} {col} {directions[current_dir]}")
print(f"Health: {health}")
````
