# Motivating Example for Version Control

You're part of a cross-functional team preparing a critical business report for a client pitch.

Your team is:

- Iterating rapidly
- Collaborating across departments
- Revising sections with sensitive, high-stakes data
- Sharing drafts via email, Teams, or shared drives

You soon have:

- ```Pitch_draft_v1.pptx```
- ```Pitch_JaneUpdated_v1.2_FINAL.pptx```
- ```Pitch_ClientEdits_FINAL_FINAL(1).pptx```

Potential problems:
- You can’t trace who changed what and when after a few iterations.
- Files get overwritten or lost.
- You waste time reconciling versions manually.
- Accountability is murky — if a mistake gets sent to the client, who made it?
- Rolling back to a “clean” version is nearly impossible without rework.

# Git: A Distributed Version Control System

### Repository (Repo)
A repository (repo) is a container. A git repo is simply the container in your project to track all changes made to files in this project. The repo allows you to access **historical** versions when needed.

### The workflow of Git

Your files 'live' in four places in Git:
- The workspace: Your working directory where code is edited **locally**
- The staging area: A temporary holding place for changes before they are committed to the **local repo**
- The local repo: The places to store committed changes **locally**
- The remote repo: Some **remote server** like GitHub to store and share code

We use Git commands to make code travel among these four places.

### Version control on your local device

Let's first make sure we have Git installed. In your terminal, type `git --version`. If Git is installed, you should see the version information.

Now follow the steps below:
1. Configure Git: Use `git config --global user.name "YourName"` and `git config --global user.email "YourEmail"` to let Git know who you are.
2. Create a working directory using `mkdir`. You can also drag and drop the folder to the workspace of your VSCode.
3. Navigate to the working directory in terminal using `cd`.
4. Create some files for us to work with. You should see the file using `ls`.
5. Use `git init` to initialize an empty **local** repo. You should see it using `ls -a`.
6. Use `git status` to see tracked and untracked files:
   - tracked files: Git knows about and are added to the repo.
   - untracked files: They are in your working directory, but not have been added to repo.
7. Let's make some changes to the test.py file.
8. Use `git add test.py` to add the file to staging area. You can use `git add .` to stage all files.
   - Staged files are ready to be committed to the repo. 
   - If you would like to **unstage** a file, use `git rm --cached <FileName>`
9. Use `git commit -m "YourMessage"` to commit your staged files to the repo. 
    - commit can be understood as a checkpoint or snapshot of your file.
    - "YourMessage" should be descriptive so that you can refer to it to understand the changes
    - The commit meesage needs to be informative and succinct. Try to answer this question when writing the message: *"**What will happen to your codebase after integrating this commit?**"*

In case you make some changes after the staging, you can compare the differences between staged file and the current file using `git diff`.

### Review Changes

To view the commits, type `git log`.

To go back to a previous version and review it, type `git checkout <commit-hash>`

If you decide to go to where you were previously, you can use `git checkout -` or `git checkout main`.

Note: `git checkout` will not revert the commits. 

### Revert Changes

Let's now see how to revert commits using Git.

Let's first make some edits on the files in our working directory, and commit the edited files to our repo.

To view the commits, type `git log`.

To revert a commit, type `git revert <commit-hash>`. Reverting to certain commits may lead to conflicts.

Use `git log` to see the commits. History is **intact** using `git revert`.

### Branch

<img src="assets/branch.svg" width="800"/> Image credit: Google image

A **branch** in Git represents an independent line of development in your project.

By default, your Git starts with the main branch or master branch.

Branches allow you to experiment new development ideas, without modifying your main codebase, avoiding conflicts or issues in the stable version of your project.

To create a new branch, type `git branch <BranchName>`. This new branch will inherit the current state of your repo.

Type `git switch <BranchName>` to switch from main to the new branch. Use `git branch` to list the braches of your project.

Try to make some changes to some files. Add the changed files to staging area, and commit those changes.

Use `git switch main` to go back to the main branch. Type `git log`. Can you see the new commits you made on `BranchName`?

### Merge Branches

Once you finish your work on one branch, you may want to integrate the changes to other branches, e.g., your stable codebase. 

To do so, use `git merge <BranchName>`. 

If you want to discard a branch, you can use `git branch -d <BranchName>`.

### Conflict Resolution

**Why Do Git Conflicts Occur?**

Git conflicts happen when two or more people/branches make changes to the same part of a file and then try to merge their work together. Git doesn't know which changes to keep, so it asks for your help to resolve the conflict.

**How to Resolve a Git Conflict?**

1. Open your code editor.
2. You'll see conflict markers in the file that looks like <<<<<<<, =======, and >>>>>>>. Everything between <<<<<<< and ======= is your changes, and everything between ======= and >>>>>>> is the other person's changes.
3. Manually decide which changes to keep or combine the changes if necessary. Remove the conflict markers and make the code look how you want it.
4. Save the file.
5. Stage the resolved file using git add.
6. Commit the changes with a message that describes what you did to resolve the conflict.
7. Finally, you can push your changes to the remote repository.

# GitHub

Git and GitHub are **different** concepts.

GitHub is a cloud platform to store your Git repos online, enabling you to share and collaborate with others. 

<img src="assets/github.png" width="800"/> Image credit: Google image

### Local vs Remote Repo

A **local** repo is those stored privately on your working device.

A **remote** repo is a version of your project stored in the server such as GitHub, or hosted by other collaborators.

The **origin** is the alias of the repo stored in GitHub. This is the naming convention.

### Connect Github with Local Repo

Versions stored in your local repo is not recognized by GitHub since we have not connected them. To establish this connection, follow the steps below:

1. Go to GitHub, login, and create a new repo.
2. Copy the url of this new repo. 
3. In your terminal, type `git remote add origin <Repo-URL>` to add the remote repo as origin.

You are now ready to work with GitHub. You can use `git remote -v` to see it.

### Git Push: Send Your Code to Remote Repo

**Git push**: This pushes your code to the origin in the same branch you are working on.

**Git push remote**: This pushes your code to arbitrary remote repo.

In your terminal, type `git push -u origin main` to send your code to GitHub.

Refresh your GitHub webpage, you should see all files in your local repo there.

### Git Pull: Fetch Code from Remote and Merge to Local Files

**Git pull < remote > < local >**: This fetches remote to the local repo and merges them.

Let's first make some update on GitHub. (**This is for demo, and is NOT recommended**)

Type `git pull origin main` to pull the changes to our main branch. 

You should see the changes sync up to our local files.

### Git Clone: Clone the Remote to Your Local

Let's navigate to a different folder. 

Use `git clone <remote-url>` to clone the remote repo to your local

You should now see the project on GitHub show up in your local

### Pull Request (PR)

A **pull request** is created when you want to merge changes from one branch into another in a repo, often involving collaboration and code review. It's typically created in a remote repo such as GitHub.

1. Let's create a new branch using `git branch -c <branch-name>`
2. Make some edits in this new branch, stage the files, and commit
3. Use `git push origin <branch-name>` to push the changes to GitHub
4. Go to the webpage of GitHub, you should see New PR. This is to request code review from other collaborators before merging
5. If code is ready to merge, then confirm merge

### Normal Workflow using GitHub: Clone, Branch, Push, PR, Merge, Pull

Here is a normal workflow when using GitHub:
1. **Clone** the repo
2. Create a new **branch** from the main or another branch
3. Make your changes and edits
4. Add the files to staging area and commit changes
5. **Push** the branch to the remote repo
6. Open a **Pull Request (PR)**
7. **Merge** the changes
8. **Pull** the merged changes to your local main branch
9. Repeat Step 2

### Git Fork: Contribute to Other's Repo

If you are not added as a collaborator or granted write access, generally you cannot push to a remote repo.

However, you can still contribute to this project. This is typically the case for developing open-source projects.

### Advanced Discussion: Git Rebase

Rebase: Reapplies commits from one branch onto another, **rewriting** the commit history to make it linear. 

Rebase moves the base of the feature branch to the tip of the target branch.

Although it cleans the commit history, but it **changes commit history**. 

**Be careful of rebase, especially for shared branches.**

Usage of rebase:

`git checkout <branch-name>`

`git rebase main`

### Rebase vs Merge

If you would like to preserve the history, use merge instead of rebase.

1. **History**: Merging retains the commit history of both branches, while rebasing rewrites the commit history to make it appear as if the changes happened sequentially.
2. **Merge Commits**: Merging creates a new merge commit, whereas rebasing does not.
3. **When to Use**: Use merging when you want to maintain the original branch's history. Use rebasing when you want a cleaner, more linear history.
